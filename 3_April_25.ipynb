```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect=CountVectorizer(max_features=1500)

X=count_vect.fit_transform(df_train.text)


X_train_count=X.toarray()
X_train_count


```

```python
from sklearn.naive_bayes import BernoulliNB

Y=df_train['target']

model=BernoulliNB()
model.fit(X,Y)
```

```python
y_pred=model.predict(X)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y,y_pred)
```

```python
print(accuracy)
```

```python
#18. Use TfidfVectorizer instead of CountVectorizer and MultinomialNB

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB


tfid_vec=TfidfVectorizer()
X_train=tfid_vec.fit_transform(df_train['text'])


X_test=tfid_vec.transform(df_test['text'])

Y=df_train['target']


model=MultinomialNB()
```

```python
model.fit(X_train,Y)
```

```python
X_train.shape, X_test.shape
```

```python
y_pred_mul=model.predict(X_test)
```

```python
y_pred_mul
```

```python
accuracy_mul=accuracy_score(df_test['target'],y_pred_mul)
```

```python
accuracy_mul
```

```python
#20. Repeat without stopwords (English)

#Removes common English stopwords (e.g., "the", "is", "and") to focus on meaningful words.

tfid_no_words=TfidfVectorizer(stop_words='english')  
X_train=tfid_no_words.fit_transform(df_train['text'])
X_test=tfid_no_words.transform(df_test['text'])

Y=df_train['target']


model=MultinomialNB()

model.fit(X_train,Y)
```

```python
X_train.shape, X_test.shape
```

```python
y_pred_no=model.predict(X_test)
```

```python
accuracy_no=accuracy_score(df_test['target'],y_pred_no)
```
